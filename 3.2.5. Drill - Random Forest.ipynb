```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
y2015 = pd.read_csv('https://www.dropbox.com/s/0so14yudedjmm5m/LoanStats3d.csv?dl=1', skipinitialspace=True, header=1)
print('Number of columns:', len(y2015.columns))
```

```python
# Convert ID and Interest Rate to numeric.
y2015['id'] = pd.to_numeric(y2015['id'], errors='coerce')
y2015['int_rate'] = pd.to_numeric(y2015['int_rate'].str.strip('%'), errors='coerce')

# Drop other columns with many unique variables
y2015.drop(['url', 'emp_title', 'zip_code', 'earliest_cr_line', 'revol_util', 'sub_grade', 'addr_state', 'desc'], 1, inplace=True)

# Remove two summary rows at the end that don't actually contain data.
y2015 = y2015[:-2]
```

```python
# Create a correlation matrix and delete closely correlated columns (> 0.5).
col_corr = set() 
corr_matrix = y2015.corr().abs()
for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if corr_matrix.iloc[i, j] >= 0.5:
            colname = corr_matrix.columns[i] 
            col_corr.add(colname)
            if colname in y2015.columns:
                del y2015[colname] 
corr_matrix
```

```python
print('Number of columns:', len(y2015.columns))
```

The number of columns dropped from the original 111 to 47 after dropping related features. The remaining columns are shown below.

```python
for i in y2015.columns:
    print(i)
```

```python
# Run the Random Forest classifier.
from sklearn import ensemble
from sklearn.model_selection import cross_val_score

rfc = ensemble.RandomForestClassifier()
X = y2015.drop('loan_status', 1)
Y = y2015['loan_status']
X = pd.get_dummies(X)
X = X.dropna(axis=1)

print('Scores:', cross_val_score(rfc, X, Y, cv=10))
print('\n')
print('Average score:', cross_val_score(rfc, X, Y, cv=10).mean())
```

The model still scored above 90% after dropping more than half of the features. The following columns, which are payment amount or outstanding principal related, were dropped during the correlation matrix run with no adverse impact to the model: out_prncp, out_prncp_inv, total_pymnt, total_pymnt_inv, installment, total_rec_prncp, and last_pymnt_amnt.
