```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Drills
df = pd.read_csv('menu.csv')
df
```

```python
# (1)
sns.factorplot(x='Calories', y='Category', data=df, kind='box', size=5, aspect=2, orient='h')
plt.title("Boxplot: McDonald's Menu Category and Calories")

sns.factorplot(x='Calories', y='Category', data=df, kind='violin', size=5, aspect=2, orient='h')
plt.title("Violinplot: McDonald's Menu Category and Calories")

sns.factorplot(x='Calories', y='Category', data=df, kind='bar', size=5, aspect=2, orient='h')
plt.title("Barplot: McDonald's Menu Category and Calories")

sns.factorplot(x='Calories', y='Category', data=df, kind='strip', size=5, aspect=2, orient='h')
plt.title("Stripplot: McDonald's Menu Category and Calories");
```

```python
# (2)
sns.lmplot(x='Calories', y='Saturated Fat', hue='Category', data=df, fit_reg=False, scatter_kws={'alpha':0.4})
plt.xlim(0,None)
plt.suptitle("Scatterplot: McDonald's Menu Calories vs. Saturated Fat");
```

```python
df_cat = (df.groupby(['Category']).agg({'Calories': 'mean', 'Saturated Fat': 'mean'}).sort_values(by='Calories', ascending=False)).reset_index()
df_cat
```

```python
df_heat = df_cat.pivot(index = 'Category', columns = 'Calories', values = 'Saturated Fat')
sns.heatmap(df_heat,annot=True, linewidths=.5);
```

```python
df1 = df[['Category','Calories','Saturated Fat','Cholesterol']]
sns.pairplot(df1, hue='Category');
```

```python
# (3)
df2 = pd.read_csv('Pokemon.csv', index_col=0, encoding='mac_turkish')
```
