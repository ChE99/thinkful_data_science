```python
# Rename the cost column.
df.rename(columns = {'PMID/PMCID': 'pmid/pmcid', 
                    'Publisher': 'publisher', 
                    'Journal title': 'journal_title', 
                    'Article title': 'article_title', 
                    'COST (£) charged to Wellcome (inc VAT when charged)': 'cost'}, inplace=True)

# Extract the digits/decimal from the cost column.
df['cost'] = df['cost'].str.extract('(\d+.\d+)', expand=True)

# Convert to numeric.
df['cost'] = pd.to_numeric(df['cost'])
```
